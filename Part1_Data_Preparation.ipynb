### 3. Are there any string or datetime columns? How to deal with them?
Use the apply() with user-defined functions (or lambda expressions) when appropriate.

```python
# We will first look at the 6 columns that are not int64 or float64 data types
df[["x2", "x3", "x14", "x25", "x26", "x29"]]
```

1. The columns **"x2"** and **"x29"** seem to be some sort of an ID, so cannot be used for any modeling. So I think we can drop them.
2. Column **"x3"** can be encoded into 1 and 0 as it is binary data (double confirmed from the df.nunique output earlier in cell 8).

```python
df = df.drop(columns=['x2'])
df = df.drop(columns=['x29'])
df['x3'] = df['x3'].map({'Male': 0, 'Female': 1})
```

3. Column "x14" seems to be integer data, just stored as object datatype. So we check further.

```python
df['x14'].unique()
```

```python
# We cast ' ' as NaN, and then convert the column
df['x14'] = df['x14'].replace(' ', np.nan)
df['x14'] = df['x14'].astype('Int64')
```

4. Columns "x25" and "x26" seem to have binary type categorical data, which can be encoded to 1 and 0

```python
df['x25'].unique()
```

```python
df['x25'] = df['x25'].map({'D_C': 0, 'L_C': 1})
```

```python
df['x26'].unique()
```

```python
df['x26'] = df['x26'].map({'Bt': 0, 'PT': 1})
```

```python
# Let us look at the data again
df.info()
```

### 4. Are there any missing data? How to deal with the missing data?

```python
# Removing unnecessary index column
df = df.drop(columns=['Unnamed: 0'])
```

```python
# Dropping fully-empty column
df = df.drop(columns=['x30'])
```

```python
# Columns x14 and x24 have nulls
df[['x14', "x24"]].isnull().sum()
```

```python
# Checking for the type/distribution of data present
df[['x14', "x24"]]
```

```python
# Filling missing values with median
df['x14'] = df['x14'].fillna(df['x14'].median())
df['x24'] = df['x24'].fillna(df['x24'].median())
```

```python
df.info()
```

### 8. Explore the continuous data columns. Try to Bin the data into some bins or quantiles.

```python
# Choosing the columns to bin
df.nunique().sort_values()
```

As only columns x5, x23 and x31 have >10 unique values, it is better to implement binning for those.

Let us go with deciles, as the the range of the data varies between the columns.

```python
cols_to_bin = df[['x5', 'x23', 'x31']]
```

```python
df_binned = df.copy()
for col in cols_to_bin:
    df_binned[col + '_qbin'] = pd.qcut(df[col], q=10,labels=False, duplicates='drop')
```

```python
df_binned[['x5', 'x5_qbin', 'x23', 'x23_qbin', 'x31', 'x31_qbin']].describe()
```

```python
df_binned = df_binned.drop(columns=cols_to_bin)
```

```python
df_binned.info()
```

### 9. Visualizations

```python
# Even though we have eliminated the highly correlated features, we can still take a look at the heatmap to confirm
plt.figure(figsize=(14, 12))
sns.heatmap(df_binned.corr(numeric_only=True), cmap='coolwarm', center=0)
plt.title("Correlation Heatmap of Features")
plt.show()
```

```python
# Boxplot for checking outliers
plt.figure(figsize=(20, 5))
sns.boxplot(data=df_binned)
plt.title("Boxplots of Features")
plt.tight_layout()
plt.show()
```

```python
# To check if the classes are imbalanced
sns.countplot(x=df_binned['y'])
plt.title("Distribution of Target Variable y")
plt.show()
```

```python
# feature vs target boxplots
features_to_plot = ['x5_qbin', 'x23_qbin', 'x31_qbin']

plt.figure(figsize=(14, 6))
for i, col in enumerate(features_to_plot):
    plt.subplot(1, 3, i+1)
    sns.boxplot(x=df_binned['y'], y=df_binned[col])
    plt.title(f'{col} vs y')
plt.tight_layout()
plt.show()
```

```python
# Distribution plots to see how the other columns that were not binned are distributed
ordinal_cols = [col for col in df_binned.columns if df_binned[col].nunique() <= 7 and col != 'y']

plt.figure(figsize=(16, 14))
for i, col in enumerate(ordinal_cols[:12]):
    plt.subplot(4, 3, i+1)
    sns.countplot(x=df[col])
    plt.title(f'Distribution of {col}')
plt.tight_layout()
plt.show()
```
